### Combine Multiple Raw CSV Files into One Dataset

This section reads all CSV files from the **raw folder**, processes them, and saves a combined dataset into the **processed folder**.

---

```python
combined_all_data = []

for file in os.listdir(raw_folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(raw_folder_path, file)

        # Read CSV
        df = pd.read_csv(file_path, encoding="utf-8-sig")

        # Select specific columns
        df_selected = df[["zilla_name", "seat_id","seat_name","center_no","center_name","legal_vote","cancel_vote","Absent_vote","total_vote","candidate_name","symbol","candidate_vote"]]   # change to your column names

        combined_all_data.append(df_selected)

# Combine all data
combined_all_df= pd.concat(combined_all_data, ignore_index=True)

 
output_file = os.path.join(processed_folder_path, "all_combined.csv")
# Save to new CSV
combined_all_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print("File created:{output_file}")
```

### Extract Unique Seat Information and Save as Separate File

This section creates a **seat-level dataset** from the combined election data.

---

```python

file_to_read = os.path.join(processed_folder_path, "all_combined.csv")
combined_df = pd.read_csv(file_to_read, encoding="utf-8-sig")
selected_seat_df = combined_df[["zilla_name", "seat_id","seat_name"]].drop_duplicates()
selected_seat_df["seat_id"] = selected_seat_df["seat_id"].astype(int)
# Reset index
df_selected_seat = selected_seat_df.reset_index(drop=True)
display(df_selected_seat)
output_file_seat = os.path.join(processed_folder_path, "seats.csv")
# Save to new CSV
selected_seat_df.to_csv(output_file_seat, index=False, encoding="utf-8-sig")
print("File created:")


```

### Extract Polling Center-Level Data and Save as Separate File

This section creates a **polling center summary dataset** from the combined election data.

---

```python
file_to_read = os.path.join(processed_folder_path, "all_combined.csv")
combined_df = pd.read_csv(file_to_read, encoding="utf-8-sig")
selected_Pooling_center_df = combined_df[[ "seat_id","center_no","center_name","legal_vote","cancel_vote","Absent_vote","total_vote"]].drop_duplicates()
# Reset index
selected_Pooling_center_df = selected_Pooling_center_df.reset_index(drop=True)



selected_Pooling_center_df["seat_id"] = selected_Pooling_center_df["seat_id"].astype(int)
selected_Pooling_center_df["center_no"] = selected_Pooling_center_df["center_no"].astype(int)
selected_Pooling_center_df["legal_vote"] = selected_Pooling_center_df["legal_vote"].astype(int)
selected_Pooling_center_df["cancel_vote"] = selected_Pooling_center_df["cancel_vote"].astype(int)


selected_Pooling_center_df["Absent_vote"] = selected_Pooling_center_df["Absent_vote"].astype(int)
selected_Pooling_center_df["total_vote"] = selected_Pooling_center_df["total_vote"].astype(int)


selected_Pooling_center_df["cast_vote"] = selected_Pooling_center_df["legal_vote"] + selected_Pooling_center_df["cancel_vote"]
selected_Pooling_center_df["cast_vote"] = selected_Pooling_center_df["cast_vote"].astype(int)

display(selected_Pooling_center_df)
output_file = os.path.join(processed_folder_path, "pooling_centres.csv")
# Save to new CSV
selected_Pooling_center_df.to_csv(output_file, index=False, encoding="utf-8-sig")
print("File created:")
```

### Extract Candidate-Level Data and Save as Separate File

This section creates a **clean candidate-level dataset** from the combined election data.

```python
file_to_read = os.path.join(processed_folder_path, "all_combined.csv")
combined_df = pd.read_csv(file_to_read, encoding="utf-8-sig")
selected_candidate_data_df = combined_df[[ "seat_id","center_no","candidate_name","symbol","candidate_vote"]].drop_duplicates()
# Reset index
selected_candidate_data_df = selected_candidate_data_df.reset_index(drop=True)

selected_candidate_data_df["seat_id"] = selected_candidate_data_df["seat_id"].astype(int)
selected_candidate_data_df["center_no"] = selected_candidate_data_df["center_no"].astype(int)

selected_candidate_data_df["candidate_vote"] = selected_candidate_data_df["candidate_vote"].astype(int)



display(selected_candidate_data_df)
output_file = os.path.join(processed_folder_path, "candidates_data.csv")
# Save to new CSV
selected_candidate_data_df.to_csv(output_file, index=False, encoding="utf-8-sig")
print("File created:")
```
